```python
# STEP 1: Install dependencies (if not already installed)
!pip install spacy
!python -m spacy download en_core_web_md  # Medium model (good balance of accuracy and speed)

# STEP 2: Import required libraries
import spacy
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

# Load spaCy English model
nlp = spacy.load("en_core_web_md")  # Use "en_core_web_lg" for a more accurate model (but slower)

# STEP 3: Upload CSV File (Run this cell and manually upload your file)
from google.colab import files
uploaded = files.upload()

# STEP 4: Read the CSV file into a list (Assuming words are in the first column)
filename = list(uploaded.keys())[0]  # Get the uploaded file name
df = pd.read_csv(filename)  # Read the CSV

# Assuming words are in the first column
column_name = df.columns[0]  # Automatically detect first column name
words = df[column_name].dropna().tolist()  # Convert non-empty values to a list

print(f"‚úÖ Loaded {len(words)} words from '{filename}'.")

# STEP 5: Convert words into vector embeddings
vectors = []
valid_words = []

for word in words:
    token = nlp(word)
    if token.has_vector:  # Some rare words might not have vectors
        vectors.append(token.vector)
        valid_words.append(word)

vectors = np.array(vectors)  # Convert to NumPy array

# STEP 6: Cluster words using KMeans
num_clusters = 10  # Change this to increase or decrease the number of categories
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
labels = kmeans.fit_predict(vectors)

# STEP 7: Organize words into clusters
clusters = {}
for label, word in zip(labels, valid_words):
    clusters.setdefault(label, []).append(word)

# STEP 8: Save results to CSV file
output_df = pd.DataFrame({'Word': valid_words, 'Cluster': labels})
output_filename = "clustered_words.csv"
output_df.to_csv(output_filename, index=False)

print(f"‚úÖ Clustered words saved as '{output_filename}'.")

# STEP 9: Print a sample of categorized words
for cluster_label, cluster_words in clusters.items():
    print(f"\nüîπ Cluster {cluster_label}:")
    print(", ".join(cluster_words[:10]))  # Show only the first 10 words in each cluster
```
